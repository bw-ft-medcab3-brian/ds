```
import pandas as pd
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/abgrss/Cannabis-NLP/master/data/df_strain_me_v2.csv')

df = df.drop(columns=['Unnamed: 0', 'cluster', 'in_ocs',
                      'lean', 'dry eyes', 'dry mouth',
                      'headache', 'Rating', 'NumReviews'])  # Dry eyes / mouth, headache side effects
                      # Getting rid of rating, keeping scaled rating - because of this, dropping number of reviews as well.
                      # Not doing the same with THC Scaled due to this likely being a THC THC-a type thing, and there are uses for each.

df = df[df.columns.drop(list(df.filter(regex='_S')))]


list(df.columns)
```

```
symp = df.columns[47:-5]  # Symptoms
df[symp] = df[symp].replace(np.nan, 0.0)  # Setting this to zero before normalization.
```

```
# Normalize our values
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df[symp] = scaler.fit_transform(df[symp])
```

```
# Replace flavor NaN with 0.

flavors = df.columns[1:47]

df[flavors] = df[flavors].replace(np.nan, 0)
```
